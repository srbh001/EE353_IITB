### Observation:
We see that many columns are **highly skewed**, with long tails, particularly in **BALANCE**, **PURCHASES**, **CASH_ADVANCE**, and **MINIMUM_PAYMENTS**. These may benefit from **logarithmic transformations** (`np.log1p`) to reduce skewness. 

Columns like **BALANCE_FREQUENCY** and **CASH_ADVANCE_FREQUENCY** are closer to binary, requiring **normalization** or no transformation. 

**Frequency features** (e.g., **PURCHASES_FREQUENCY**) may benefit from **square root transformations** (`np.sqrt`) to smooth the distribution.

```python

transformed_df = df.copy()

# Columns that do not require transformations
untouched_cols = [
    'TENURE',         # Categorical or limited range
    'PRC_FULL_PAYMENT'  # Bimodal distribution
]

# Columns for logarithmic transformation
log_transform_cols = [
    'BALANCE', 'PURCHASES', 'ONEOFF_PURCHASES', 
    'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 
    'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 
    'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS'
]

# Columns for square root (power) transformation
sqrt_transform_cols = [
    'PURCHASES_INSTALLMENTS_FREQUENCY', 
    'CASH_ADVANCE_FREQUENCY'
]


# Logarithmic transformation (log1p handles zeros safely)
for col in log_transform_cols:
    transformed_df[col] = np.log1p(transformed_df[col])

# Square root transformation
for col in sqrt_transform_cols:
    transformed_df[col] = np.sqrt(transformed_df[col])


# Ensure untouched columns remain unchanged
transformed_df[untouched_cols] = df[untouched_cols]

transformed_df.head()
```

```python
# Determine grid dimensions
cols_per_row = 3
n_cols = cols_per_row
n_rows = -(-len(df.columns) // n_cols)  # Ceiling division

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 4), gridspec_kw={'wspace': 0.4, 'hspace': 0.6})

# Flatten axes for easy indexing
axes = axes.flatten()

# Plot each column
for i, col in enumerate(transformed_df.columns):
    sns.histplot(transformed_df[col], ax=axes[i], kde=True)
    axes[i].set_title(f'Distribution: {col}')

# Turn off unused subplots
for j in range(len(transformed_df.columns), len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.savefig('plot.png')
plt.show()
```

```python
# Function to drop highly correlated columns
def drop_highly_correlated(df, threshold=0.8):
    corr_matrix = df.corr(numeric_only=True).abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    return df.drop(columns=to_drop), to_drop

# Drop columns
df, dropped_columns = drop_highly_correlated(df, threshold=0.8)



# Results
print("Dropped Columns:", dropped_columns)
print("Filtered DataFrame:")
df.head()
```

## Clustering 

### K-Means

```python

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
# Normalize data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
# [1] : Find optimal number of clusters using Elbow Method
inertia = []
range_n_clusters = range(1, 11)
for k in range_n_clusters:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.figure(figsize=(10, 8))
plt.plot(range_n_clusters, inertia, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()

# Silhouette Analysis
silhouette_scores = []
for k in range(2, 11):  # Silhouette score needs at least 2 clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(scaled_data)
    silhouette_avg = silhouette_score(scaled_data, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot Silhouette Analysis
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Analysis')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.show()

# Train K-Means with optimal number of clusters
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Output results
print(f"Optimal number of clusters (k): {optimal_k}")
print(df['Cluster'].value_counts())
```

```python
## Visualizing this cluster using t-sne
from sklearn.manifold import TSNE
# Perform t-SNE dimensionality reduction
tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
tsne_embedding = tsne.fit_transform(scaled_data)

# Add t-SNE results to DataFrame for visualization
df['tSNE_1'] = tsne_embedding[:, 0]
df['tSNE_2'] = tsne_embedding[:, 1]

# Plot t-SNE embedding
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x='tSNE_1', y='tSNE_2', 
    hue='Cluster', 
    palette='viridis', 
    data=df, 
    s=50, alpha=0.8
)
plt.title(f't-SNE Visualization of Clusters (k={optimal_k})')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend(title='Cluster')
plt.show()
```

### DB-SCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python


def train_dbscan(scaled_data, eps_values, min_samples_values):
    results = []
    for eps in eps_values:
        for min_samples in min_samples_values:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels = dbscan.fit_predict(scaled_data)

            # Count unique clusters (ignoring noise points labeled as -1)
            n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

            # Calculate Silhouette Score (only if more than 1 cluster is formed)
            if n_clusters > 1:
                silhouette_avg = silhouette_score(scaled_data, labels)
            else:
                silhouette_avg = None

            results.append({
                'eps': eps,
                'min_samples': min_samples,
                'n_clusters': n_clusters,
                'silhouette_score': silhouette_avg
            })
    return results

# Define parameter ranges to explore
eps_values = [0.1, 0.3, 0.5, 0.7, 1.0, 1.5, 2.0]
min_samples_values = [3, 4, 5, 6, 7, 8, 9, 10]

# Train and collect results
dbscan_results = train_dbscan(scaled_data, eps_values, min_samples_values)

# Display results
for res in dbscan_results:
    print(f"ε: {res['eps']}, minPts: {res['min_samples']}, "
          f"Clusters: {res['n_clusters']}, Silhouette: {res['silhouette_score']}")

# Find configurations with the same number of clusters as K-Means
matching_results = [res for res in dbscan_results if res['n_clusters'] == optimal_k]

print("Matching Configurations with K-Means Clusters:")
for res in matching_results:
    print(f"ε: {res['eps']}, minPts: {res['min_samples']}, "
          f"Silhouette: {res['silhouette_score']}")
```

```python

best_eps = matching_results[0]['eps'] 
best_min_samples = matching_results[0]['min_samples']  

# Train DBSCAN with the best parameters
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
dbscan_labels = dbscan.fit_predict(scaled_data)

# DBSCAN visualization
dbscan = DBSCAN(eps=1.5, min_samples=9).fit(scaled_data)
tsne_dbscan = TSNE(n_components=2, random_state=42).fit_transform(scaled_data)

plt.figure(figsize=(12, 8))
plt.scatter(tsne_dbscan[:, 0], tsne_dbscan[:, 1], c=dbscan.labels_, cmap='viridis')
plt.title("t-SNE Embedding of DBSCAN Clusters")
plt.colorbar()
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error
```

```python

df = transformed_df.copy()
# Apply PCA
pca = PCA()
pca.fit(df)

# Plot Variance Explained vs. PCA Dimensions
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()

# Reconstruct the data with various numbers of components and calculate MSE
mse_values = []
components_range = range(1, len(explained_variance_ratio) + 1)

for n_components in components_range:
    # Apply PCA with 'n_components'
    pca = PCA(n_components=n_components)
    pca_transformed = pca.fit_transform(df)
    reconstructed_data = pca.inverse_transform(pca_transformed)
    
    # Compute Mean Squared Error (MSE) between original and reconstructed data
    mse = mean_squared_error(df, reconstructed_data)
    mse_values.append(mse)

# Find the optimal number of components (the "elbow" where MSE starts leveling off)
optimal_components = components_range[np.argmin(mse_values)]

# Plot MSE vs. Number of Components
plt.figure(figsize=(10, 6))
plt.plot(components_range, mse_values, marker='o', linestyle='--')

# Highlight the optimal number of components
plt.axvline(optimal_components, color='r', linestyle='--')
plt.text(optimal_components, min(mse_values), f'Optimal {optimal_components} Components', color='r', ha='left')

plt.title('MSE vs. Number of PCA Components')
plt.xlabel('Number of Components')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()
```

### PCA Analysis Summary

- **Cumulative Explained Variance**:  
  The top graph shows that the cumulative explained variance plateaus after 17 components, capturing almost all the variance in the data. This suggests that 17 components are sufficient to represent the data effectively.

- **Mean Squared Error (MSE)**:  
  The bottom graph demonstrates a significant drop in MSE as the number of components increases, with minimal improvement after 17 components. This indicates that 17 components provide an optimal balance between dimensionality reduction and reconstruction error.

### Conclusion  
The ideal PCA model is achieved with **17 components**, as it balances variance explanation and error minimization.
