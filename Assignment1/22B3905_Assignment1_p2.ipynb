```python

# print the names of columns which do not have type float to delete them later
for col in df:
    if  df.dtypes[col] != "float64":
        print(col,  df.dtypes[col])
```

```python
# List of columns to convert to float
cols_str = [
    'Median age of initiation of Cigarette (in years)',
    'Median age of initiation of Bidi (in years)',
    'Median age of initiation of smokeless tobacco (in years)',
    'Bought Cigarette as individual sticks in past 30 days    (%)',
    'Students who saw anyone using tobacco on mass media in past 30 days  (%)'
]

# Convert columns to float, dropping rows with invalid data
for col in cols_str:
    # Convert column to numeric, forcing errors to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows where any of the specified columns have NaN values
df_cleaned = df.dropna(subset=cols_str)


print("DataFrame columns after cleaning:")
print(df_cleaned.columns)
print("Number of rows before cleaning:", len(df))
print("Number of rows after cleaning:", len(df_cleaned))
```

```python
df['Current_Smoke'] = (
    df['Current tobacco users (%)'] +
    df['Current tobacco smokers (%)'] +
    df['Current cigarette users (%)'] +
    df['Current bidi users (%)'] +
    df['Current smokeless tobacco users (%)']
)

# Students Notices Anti-Tobacco metrics
df['Students_Notices_Anti_Tobacco'] = (
    df['Students who noticed anti-tobacco messages anywhere in past 30 days    (%)'] +
    df['Students who noticed anti-tobacco messages in mass media in past 30 days    (%)'] +
    df['Students who noticed anti-tobacco messages at sporting,  fairs, concerts, community events or social gatherings in past 30 days   (%)'] +
    df['Students who noticed health warnings on any tobacco product/cigarette packages in past 30 days       (%)']
)

# Students Notices Ads metrics
df['Students_Notices_Ads'] = (
    df['Students who saw tobacco advertisements anywhere in past 30 days  (%)'] +
    df['Students who saw anyone using tobacco on mass media in past 30 days  (%)'] +
    df['Students who noticed cigarette advertisements/promotions at point of sale in past 30 days (%)']
)

# In Favor of Tobacco-Free Schools metrics
df['In_Favor_Tobacco_Free_Schools'] = (
    df['Students who favoured ban on smoking inside enclosed public places (%)'] +
    df['Students who favoured ban on smoking at outdoor public places (%)'] +
    df['School heads aware of COTPA, 2003  (%)'] +
    df['Schools authorized by the state government to collect fine for violation under Section-6 of the COTPA, 2003  (%)'] +
    df['Schools followed ‘tobacco-free school’ guidelines  (%)'] +
    df['Schools aware of the policy for displaying ‘tobacco-free school’ board  (%)']
)

# Select only the new columns along with the state and area information
new_df = df[['State/UT', 'Area', 'Current_Smoke', 'Students_Notices_Anti_Tobacco', 'Students_Notices_Ads', 'In_Favor_Tobacco_Free_Schools']]

# Save the new DataFrame to a CSV file if needed
new_df.to_csv('aggregated_tobacco_data.csv', index=False)

new_df.head()
```

# Formulate and test the hypothesis



```python
from scipy import stats
import seaborn as sns
```


### Null Hypothesis (H0):
There is no significant relationship between the variables related to school support for tobacco-free policies and smoking-related data. Specifically, variables such as school favoring a ban on smoking or being aware of COTPA do not significantly impact smoking-related metrics.

### Alternative Hypothesis (H1):
There is a significant relationship between the variables related to school support for tobacco-free policies and smoking-related data. Specifically, variables such as school favoring a ban on smoking or being aware of COTPA significantly impact smoking-related metrics.

### Analysis and Test


```python
corr_matrix = new_df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1,
            cbar_kws={'shrink': .8})
plt.title('Correlation Heatmap')
plt.show()
```

```python
#[1]
threshold = new_df['In_Favor_Tobacco_Free_Schools'].median()

_, p_value_smoke = stats.ttest_ind(new_df[new_df['In_Favor_Tobacco_Free_Schools'] > threshold]['Current_Smoke'],
                                  new_df[new_df['In_Favor_Tobacco_Free_Schools'] <= threshold]['Current_Smoke'])

print(f"P-value for Current_Smoke: {p_value_smoke}")
```

## Conclusion from the test

We conducted a Mann-Whitney U test to compare the levels of current smoking between two groups of schools: those that favor a smoking ban or are aware of COTPA, and those that do not. 
The p-value obtained from the test was approximately _0.399_. 
This value is greater than the commonly used significance level of 0.05. Therefore, we fail to reject the null hypothesis, suggesting that there is no significant difference in current smoking rates between the two groups. 

Hence,  school policies and awareness regarding tobacco-free environments do not appear to significantly impact smoking rates among students.
